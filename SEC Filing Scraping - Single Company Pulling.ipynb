```python
#import libraries
import requests
from bs4 import BeautifulSoup

#define base url
base_url = r"https://www.sec.gov/Archives/edgar/data/"

#define CIK number for Company Search: GOLDMAN SACHS
cik_num = '/886982/'

#creating a filings url
filings_url = base_url + cik_num + "/index.json"

#request url
content = requests.get(filings_url)
decode_content = content.json()

#getting a single filing number
filing_number = decode_content['directory']['item'][0]['name']

#define filing number url
filing_url = base_url + cik_num + filing_number + "/index.json"

#request url
content = requests.get(filing_url)
document_content = content.json()

for document in document_content['directory']['item']:
    if document['type'] != 'image2.gif':
        document_name = document['name']
        document_url = base_url + cik_num + filing_number + '/' + document_name
        print(document_url)
```

```python
#define base url
base_url = r"https://www.sec.gov/Archives/edgar/data/"

#define CIK number for Company Search: GOLDMAN SACHS
cik_num = '/886982/'

#creating a filings url
filings_url = base_url + cik_num + "/index.json"

#request url
content = requests.get(filings_url)
decode_content = content.json()

#get multiple filings
for filing in decode_content['directory']['item']:
    
    #define each filing number
    filing_num = filing['name']
    filing_url = base_url + cik_num + filing_num + "/index.json"
    
    #request url
    content = requests.get(filing_url)
    document_content = content.json()
    
    for document in document_content['directory']['item']:
        document_name = document['name']
        document_url = base_url + cik_num + filing_num + '/' + document_name
        print(document_url)
    
```
